## Define YOLO paths in `data.yaml` 


The `data.yaml` file is crucial for YOLO model training as it contains paths to the dataset and other configurations. Below is an example of what the `data.yaml` file should include and where to save it.

### Example `data.yaml` File
```
%yaml

Path to the training and validation datasets
`train: /path/to/train/dataset val: /path/to/val/dataset`

Number of classes in a default yolo dataset used to train e.g. yolo11n-seg.pt
nc: 80

Class names
names: ['class1', 'class2', ..., 'class80']
```

### Notes on `data.yaml`

- **train**: Path to the training dataset.
- **val**: Path to the validation dataset.
- **nc**: Number of classes in the dataset.
- **names**: List of class names.

### Saving `data.yaml`

Save the `data.yaml` file in a directory accessible by your training script. For example, you can save it in the same directory as your training script or in a dedicated configuration directory.

### Example Directory Structure
```
/project ├── data.yaml ├── train.py └── datasets ├── train └── val
```

In this structure:  
- `data.yaml` is saved in the `/project` directory.
- `train.py` is your training script.
- `datasets` directory contains the training and validation datasets.

### Using `data.yaml` in Training Script

In your training script, you can reference the `data.yaml` file as shown below:
```
from ultralytics import YOLO

# Initialize the YOLO model
model = YOLO('yolo11n-seg.pt')

# Train the model using the data.yaml file
model.train(data='data.yaml', epochs=100, project='/path/to/specific/project')
```

This setup ensures that the YOLO model uses the paths and configurations defined in the `data.yaml` file for training.

````python
def create_yaml(output_yaml_path, train_images_dir, val_images_dir, test_images_dir, nc=1):
    # Assuming all categories are the same and there is only one class, 'Nuclei'
    names = ["Nuclei"]

    yaml_data = {
                'names': names,
                'nc': nc,  # Number of classes
                'train': train_images_dir,
                'val': val_images_dir,
                'test': test_images_dir 
                }

    # Write the dictionary to a YAML file
    with open(output_yaml_path, 'w') as file:
        yaml.dump(yaml_data, file, default_flow_style=False)

    return yaml_data        
````
